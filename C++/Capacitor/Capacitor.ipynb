```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Ex(q, x, y, X, Y):
    r = np.sqrt((X-x)**2 + (Y-y)**2)
    return q*(X-x)/r

def Ey(q, x, y, X, Y):
    r = np.sqrt((X-x)**2 + (Y-y)**2)
    return (q*(Y-y))/r
    
```

```python
q = 2.
xq = 1.0
```

```python
dq = 0.1
Lq = 4
nq = 10*Lq -1
```

```python
x = np.linspace(-10,10)
y = np.linspace(-10,10)

X, Y = np.meshgrid(x,y)
```

```python
Exx = np.zeros(X.shape)
Eyy = np.zeros(Y.shape)
```

```python
for k in range(0,nq+2):
    Exx += Ex(q, xq, -2. + k*dq, X, Y) + Ex(-q, -xq, -2. + k*dq, X, Y)
    Eyy += Ey(q, xq, -2. + k*dq, X, Y) + Ey(-q, -xq, -2. + k*dq, X, Y)

E = np.sqrt(Exx**2 + Eyy**2)
```

```python
plt.quiver(X, Y, Exx, Eyy, E)
plt.plot([-1,-1],[-2,2])
plt.plot([1,1],[-2,2])
```
