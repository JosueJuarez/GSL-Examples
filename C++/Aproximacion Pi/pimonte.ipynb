```python
from numpy import random
import numpy as np
import math
import matplotlib.pyplot as plt
from statsmodels.tools import sequences
```

```python
def random_points(N):
    x_circle = []
    y_circle = []
    x = []
    y = []

    for i in range(N):
        x.append(random.uniform(0, 1))
        y.append(random.uniform(0, 1))

    for j in range(len(x)):
        if ((x[j]) ** 2 + (y[j] ) ** 2 <= 1):
            x_circle.append(x[j])
            y_circle.append(y[j])

    pi = 4 * len(x_circle) / N
    return pi

pi_approx = random_points(1000000)
pi_real = np.pi
error_abs = abs((pi_real - pi_approx)*100/pi_real)
print('Pi estimado: ', pi_approx)
print('Pi real: ', pi_real)
print('Error absoluto%: ', error_abs)
val_pi = []
x = range(1, 1000)

for N in x:
    val_pi.append(random_points(N))

plt.plot(x, val_pi)
plt.plot([1, 1000], [pi_real, pi_real])
plt.show()
```
