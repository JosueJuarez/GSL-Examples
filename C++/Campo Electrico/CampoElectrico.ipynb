```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Ex(q, x, X, Y):
    r = np.sqrt((x-X)**2 + Y**2)
    return (q*(X-x))/r**3

def Ey(q, x, X, Y):
    r = np.sqrt((x-X)**2 + Y**2)
    return (q*Y)/r**3
    
```

```python
q = 2.
q1x = 1.0
```

```python
x = np.linspace(-10,10)
y = np.linspace(-10,10)

X, Y = np.meshgrid(x,y)
```

```python
Exx = Ex(q,q1x, X,Y) + Ex(-q,-q1x, X,Y)
Eyy = Ey(q,q1x, X,Y) + Ey(-q,-q1x, X,Y)

E = np.sqrt(Exx**2 + Eyy**2)
```

```python
E
```

```python
plt.quiver(X, Y, Exx, Eyy, E)
```
